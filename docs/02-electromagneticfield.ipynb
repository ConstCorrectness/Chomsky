# Energy is carried by the electromagnetic field:

The energy density $u$ of the electromagnetic field is given by the equation:
$$
u = \frac{1}{2} \left( \varepsilon_0 E^2 + \frac{B^2}{\mu_0} \right)
$$

where:
- $\varepsilon_0$ is the permittivity of free space,
- $E$ is the electric field strength,
- $B$ is the magnetic field strength,
- $\mu_0$ is the permeability of free space.

The total energy $U$ stored in a volume $V$ can be calculated by integrating the energy density over that volume:
$$
U = \int_V u \, dV = \int_V \frac{1}{2} \left( \varepsilon_0 E^2 + \frac{B^2}{\mu_0} \right) dV
$$


The tangential vector $\mathbf{S}$ describes the directional energy flux (the rate of energy transfer per unit area) of the electromagnetic field:
$$
\mathbf{S} = \mathbf{E} \times \mathbf{H}
$$

where:
- $\mathbf{E}$ is the electric field vector,

```python
import matplotlib.pyplot as plt
import numpy as np


fig, ax = plt.subplots(figsize=(6, 6))

# wire
ax.plot(0, 0, 'ro', markersize=12, label='Wire (current into page)')

# Magnetic field
theta = np.linspace(0, 2 * np.pi, 100)
radii = [0.3, 0.5, 0.7]
for r in radii:
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    ax.plot(x, y, 'b--', alpha=0.6)

# Electric field arrows radially out
for angle in np.linspace(0, 2*np.pi, 16):
    x = 0.9 * np.cos(angle)
    y = 0.9 * np.sin(angle)
    dx = 0.3 * np.cos(angle)
    dy = 0.3 * np.sin(angle)
    ax.arrow(x, y, dx, dy, head_width=0.05, head_length=0.05, fc='r', ec='r')

# (tangential: E × B direction)
for angle in [np.pi/4, 3*np.pi/4, 5*np.pi/4, 7*np.pi/4]:
    x = 1.2 * np.cos(angle)
    y = 1.2 * np.sin(angle)
    dx = -0.3 * np.sin(angle)
    dy = 0.3 * np.cos(angle)
    ax.arrow(x, y, dx, dy, head_width=0.05, head_length=0.05, fc='g', ec='g')

ax.text(-1.8, 1.7, "Blue dashed: Magnetic field (B)", color='blue', fontsize=10)
ax.text(-1.8, 1.5, "Red arrows: Electric field (E)", color='red', fontsize=10)
ax.text(-1.8, 1.3, "Green arrows: Poynting vector (S = E × B)", color='green', fontsize=10)


ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)

ax.set_aspect('equal')
ax.axis('off')
ax.set_title("Energy Flow Around a Current-Carrying Wire")
plt.tight_layout()
plt.show()
```
